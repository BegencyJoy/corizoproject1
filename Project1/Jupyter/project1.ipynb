 # 11. In your own words, list some advantages and disadvantages of BMI, waist-to-height ratio, and waist-to-hip ratio

<div style="text-align: left;">1. Body Mass Index (BMI)</div>
<div style="text-align: left;">Definition:</div>
              <center> BMI is calculated as weight (kg) divided by height squared (m)</center>
                                  <center> BMI=Weight(kg)/Height(m)</center>
<div style="text-align: left;">Advantages:</div>
<div style="text-align: left;">*Simple and Easy to Calculate</div>
<div style="text-align: left;">*Requires only height and weight measurements</div>
<div style="text-align: left;">*Commonly used in healthcare and research</div>
<div style="text-align: left;">*Global standard for categorizing underweight, normal weight, overweight, and obesity.</div>
<div style="text-align: left;">*Useful for Population-Level Analysis</div>
<div style="text-align: left;">*Effective for identifying trends in obesity or malnutrition across large groups.</div>
<div style="text-align: left;">* Higher BMI is associated with increased risks of chronic diseases like diabetes, heart disease, and hypertension</div>.
<div style="text-align: left;">Disadvantages:</div>
<div style="text-align: left;">*Ignores Body Composition</div>
<div style="text-align: left;">*Cannot distinguish between fat mass and muscle mass.</div>
<div style="text-align: left;">*Athletes with high muscle mass may be classified as overweight or obese.</div>
<div style="text-align: left;">*Does not account for visceral fat (fat around organs), which is more dangerous than subcutaneous fat.</div>
<div style="text-align: left;">*Less Accurate for Certain Groups</div>
 <div style="text-align: left;">*May not be suitable for children, elderly individuals, or people with different body types and ethnicities.</div>
 <div style="text-align: left;">*False Positives and Negatives</div>
<div style="text-align: left;">*People with a "normal" BMI can still have high body fat percentages, leading to underdiagnosis of obesity-related risks.</div>


<div style="text-align: left;">2. Waist-to-Height Ratio (WHtR)</div>
<div style="text-align: left;">Definition:</div>
                 <center>WHtR is calculated as waist circumference divided by height</center>
                           <center> WHtR=WaistCircumference/Height</center>
<div style="text-align: left;">Advantages:</div>
<div style="text-align: left;">*Considers Central Obesity</div>
<div style="text-align: left;">*Focuses on fat distribution in the abdominal area, which is strongly linked to health risks.</div>
<div style="text-align: left;">*Better Predictor of Health Risks</div>
<div style="text-align: left;">*More accurate than BMI for identifying risks of cardiovascular diseases, diabetes, and metabolic syndrome.</div>
<div style="text-align: left;">*Applicable Across Populations</div>
<div style="text-align: left;">*Works for all ages, genders, and ethnicities.</div>
<div style="text-align: left;">*Easy to Measure</div>
<div style="text-align: left;">*Only waist circumference and height are needed, both of which are simple to measure.</div>
<div style="text-align: left;">Disadvantages:</div>
<div style="text-align: left;">*Measurement Errors</div>
<div style="text-align: left;">*Waist circumference can be inconsistently measured depending on technique and body posture.</div>
<div style="text-align: left;">*Less Familiar in Clinical Use</div>
<div style="text-align: left;">*Not as widely used as BMI, so thresholds for health risks are less standardized.</div>
<div style="text-align: left;">*Ignores Other Body Fat Areas</div>
<div style="text-align: left;">*While focusing on abdominal fat, it ignores fat in other areas that may also contribute to health risks.</div>

<div style="text-align: left;">3. Waist-to-Hip Ratio (WHR)</div>
<div style="text-align: left;">Definition:</div>
                   <center> WHR is calculated as waist circumference divided by hip circumference</center>
                                      <center>WHR=WaistCircumference/HipCircumference</center>
<div style="text-align: left;">Advantages:</div>
<div style="text-align: left;">*Assesses Fat Distribution</div>
<div style="text-align: left;">*Distinguishes between "apple-shaped" (abdominal fat) and "pear-shaped" (hip fat) body types.</div>
<div style="text-align: left;">*Central obesity (apple shape) is more harmful than lower body fat (pear shape).</div>
<div style="text-align: left;">*Strong Indicator of Health Risks</div>
<div style="text-align: left;">*High WHR is associated with increased risks of heart disease, diabetes, and stroke.</div>
<div style="text-align: left;">*Useful for Both Genders</div>
<div style="text-align: left;">*Helps identify health risks in men and women, regardless of BMI.</div>
<div style="text-align: left;">Disadvantages:</div>
<div style="text-align: left;">*Measurement Challenges</div>
<div style="text-align: left;">*Accurate measurement of waist and hip circumference requires skill and consistency.</div>
<div style="text-align: left;">*Does Not Consider Overall Body Fat</div>
<div style="text-align: left;">*Focuses only on relative fat distribution and ignores total body fat percentage.</div>
<div style="text-align: left;">*Less Accurate in Certain Groups</div>
<div style="text-align: left;">*May not perform as well for very lean or muscular individuals.</div>
<div style="text-align: left;">*Not as Commonly Used</div>
<div style="text-align: left;">*Less standardized in clinical practice compared to BMI.</div>


 # 12. Print out the standardised body measurements for the 5 persons with the lowest BMI and the 5 persons with the 5 highest BMI (e.g., call print for a subset of zfemale comprised of 10 chosen rows as determined by a call to numpy.argsort). Interpret the results.

```python
female_data = pd.read_csv('female_bmx_data_with_ratios.csv')
#Load the female dataset and compute BMI
fweight = female_data['Female_weight']  # Column for weight in kg
fheights = female_data['Female_Height'] # Assuming 'Height' column is in meter 

female_bmi=fweight /(fheights ** 2)

#Standardize the dataset (z-scores)
zfemale = (female_data - female_data.mean()) / female_data.std()

# Find the indices of the 5 lowest and 5 highest BMIs
sorted_indices = np.argsort(female_bmi)  # Sort BMI in ascending order
lowest_bmi_indices = sorted_indices[:5]  # First 5 indices (lowest BMI)
highest_bmi_indices = sorted_indices[-5:]  # Last 5 indices (highest BMI)

# Step 4: Extract rows from the zfemale matrix
lowest_bmi_rows = zfemale.iloc[lowest_bmi_indices]
highest_bmi_rows = zfemale.iloc[highest_bmi_indices]

# Step 5: Combine and print the results
print("Standardized Body Measurements for 5 Lowest BMI:")
print(lowest_bmi_rows)

print("\nStandardized Body Measurements for 5 Highest BMI:")
print(highest_bmi_rows)
```

<h2>observations:</h2>
<div style="text-align: left;">Participants with the lowest BMI have smaller body dimensions, lower waist circumferences, and proportions that indicate minimal fat storage.</div>
<div style="text-align: left;">Participants with the highest BMI exhibit much larger body measurements, especially around the waist and hips, reflecting greater fat accumulation.</div>
<div style="text-align: left;">The Waist-to-Height Ratio highlights a key difference: higher BMI individuals tend to have disproportionately large waists relative to their height, which is a known indicator of central obesity.</div>
